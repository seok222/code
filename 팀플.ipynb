```python
#ICT 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

df = pd.read_csv("C:/Users/samsung/OneDrive/ //ICT__20221122201925(1).csv", encoding="utf-8")

#
ln00_sub = df.loc[0][1:]

ln00_sub = ln00_sub.astype(str)

ln00 = df.loc[2][1:]

ln01 = ln00.astype(int)

matplotlib.rcParams['font.family'] ='Malgun Gothic'

matplotlib.rcParams['axes.unicode_minus'] =False

print(ln00)

#   
ln000_sub = df.loc[0][1:]

ln000_sub = ln000_sub.astype(str)

ln000 = df.loc[3][1:]

ln000 = ln000.astype(int)

matplotlib.rcParams['font.family'] ='Malgun Gothic'

matplotlib.rcParams['axes.unicode_minus'] =False

print(ln000)

fig = plt.figure(figsize=(10,10)) 
fig.set_facecolor('white') 
ax = fig.add_subplot()

ax.plot(ln00_sub, ln00, color = "red", marker = "o", linestyle = "solid", label='')
ax.plot(ln000_sub, ln000, color = "green", marker = "o", linestyle = "solid", label='   ')

ax.legend() 

plt.title('ICT ',fontsize=20) 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

#  
df = pd.read_csv("C:/Users/samsung/OneDrive/ //ICT__20221122201925(1).csv", encoding="utf-8")

ln00_sub = df.loc[0][1:]

ln00_sub = ln00_sub.astype(str)

ln00 = df.loc[2][1:]

ln00 = ln00.astype(int)

matplotlib.rcParams['font.family'] ='Malgun Gothic'

matplotlib.rcParams['axes.unicode_minus'] =False

print(ln00)

# 
df = pd.read_csv("C:/Users/samsung/OneDrive/ //_20221122201954(1).csv", encoding="utf-8")

ln01_sub = df.loc[0][1:]

ln01_sub = ln01_sub.astype(str)

ln01 = df.loc[1][1:]

ln01 = ln01.astype(int)

matplotlib.rcParams['font.family'] ='Malgun Gothic'

matplotlib.rcParams['axes.unicode_minus'] =False

print(ln01)

#  
df = pd.read_csv("C:/Users/samsung/OneDrive/ //___20221122202006(1).csv", encoding="utf-8")

ln02_sub = df.loc[0][1:]

ln02_sub = ln02_sub.astype(str)

ln02 = df.loc[1][1:]

ln02 = ln02.astype(int)

matplotlib.rcParams['font.family'] ='Malgun Gothic'

matplotlib.rcParams['axes.unicode_minus'] =False

print(ln02)

fig = plt.figure(figsize=(10,10)) 
fig.set_facecolor('white') 
ax = fig.add_subplot()
ax.get_yaxis().get_major_formatter().set_scientific(False)

ax.plot(ln00_sub, ln00, color = "red", marker = "o", linestyle = "solid")
ax.plot(ln01_sub, ln01, color = "green", marker = "o", linestyle = "solid")
ax.plot(ln02_sub, ln02, color = "blue", marker = "o", linestyle = "solid")

ax.legend() 

plt.title('  ',fontsize=20) 
plt.show()
```
