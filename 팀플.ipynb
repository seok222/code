```python
#ICT산업 인력현황
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

df = pd.read_csv("C:/Users/samsung/OneDrive/바탕 화면/텀프/ICT산업_인력현황_20221122201925(1).csv", encoding="utf-8")

#정보통신방송기기업
ln00_sub = df.loc[0][1:]

ln00_sub = ln00_sub.astype(str)

ln00 = df.loc[2][1:]

ln01 = ln00.astype(int)

matplotlib.rcParams['font.family'] ='Malgun Gothic'

matplotlib.rcParams['axes.unicode_minus'] =False

print(ln00)

#소프트웨어 개발 및 제작업
ln000_sub = df.loc[0][1:]

ln000_sub = ln000_sub.astype(str)

ln000 = df.loc[3][1:]

ln000 = ln000.astype(int)

matplotlib.rcParams['font.family'] ='Malgun Gothic'

matplotlib.rcParams['axes.unicode_minus'] =False

print(ln000)

fig = plt.figure(figsize=(10,10)) 
fig.set_facecolor('white') 
ax = fig.add_subplot()

ax.plot(ln00_sub, ln00, color = "red", marker = "o", linestyle = "solid", label='정보통신방송기기업')
ax.plot(ln000_sub, ln000, color = "green", marker = "o", linestyle = "solid", label='소프트웨어 개발 및 제작업')

ax.legend() 

plt.title('ICT산업 인력현황',fontsize=20) 
plt.show()
```
